```python
# python packages to install or import
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline

# for data import or analysis
import pandas as pd
import seaborn as sns
```

```python
# loading the data into the workspace

table_video1 = pd.read_csv('data/18.01.11_BR_videos.csv', error_bad_lines=False)
table_video2 = pd.read_csv('data/18.01.11_CA_videos.csv', error_bad_lines=False)
table_video3 = pd.read_csv('data/18.01.11_DE_videos.csv', error_bad_lines=False)
table_video4 = pd.read_csv('data/18.01.11_FR_videos.csv', error_bad_lines=False)
table_video5 = pd.read_csv('data/18.01.11_GB_videos.csv', error_bad_lines=False)
table_video6 = pd.read_csv('data/18.01.11_IN_videos.csv', error_bad_lines=False)
table_video7 = pd.read_csv('data/18.01.11_JP_videos.csv', error_bad_lines=False)
table_video8 = pd.read_csv('data/18.01.11_KR_videos.csv', error_bad_lines=False)
table_video9 = pd.read_csv('data/18.01.11_MX_videos.csv', error_bad_lines=False)
table_video10 = pd.read_csv('data/18.01.11_RU_videos.csv', error_bad_lines=False)
table_video11 = pd.read_csv('data/18.01.11_US_videos.csv', error_bad_lines=False)

table_videos = pd.concat([table_video1, table_video2, table_video3, table_video4, table_video5, 
                          table_video6, table_video7, table_video8, table_video9, table_video10, table_video11])
```

```python
# Categorize through category id
category_id_map = { 1:'Film & Animation', 
                    2: 'Autos & Vehicles',
                    10: 'Music',
                    15: 'Pets & Animals',
                    17: 'Sports',
                    18: 'Short Movies',
                    19: 'Travel & Events',
                    20: 'Gaming',
                    21: 'Videoblogging',
                    22: 'People & Blogs',
                    23: 'Comedy',
                    24: 'Entertainment',
                    25: 'News & Politics',
                    26: 'Howto & Style',
                    27: 'Education',
                    28: 'Science & Technology',
                    29: 'Nonprofits & Activism',
                    30: 'Movies',
                    31: 'Anime/Animation',
                    32: 'Action/Adventure',
                    33: 'Classics',
                    34: 'Comedy',
                    35: 'Documentary',
                    36: 'Drama',
                    37: 'Family',
                    38: 'Foreign',
                    39: 'Horror',
                    40: 'Sci-Fi/Fantasy',
                    41: 'Thriller',
                    42: 'Shorts',
                    43: 'Shows',
                    44: 'Trailers'}
```

```python
# rename category id

table_videos_cat_id = table_videos.groupby('categoryId').count()['title']
table_videos_rename = table_videos_cat_id.rename(category_id_map)

# plot video counts in each category
ax = table_videos_rename.plot(kind='bar',title='Video Categories by their Count',color='green',figsize=(10,5))
ax.set_xlabel('Category')
ax.set_ylabel('Count')

plt.tight_layout()
plt.savefig('video_trends_all.png', dpi=300)
```

```python
# plot video counts in each category for US videos

table_videos11_cat_id = table_video11.groupby('categoryId').count()['title']
table_videos11_rename = table_videos11_cat_id.rename(category_id_map)

ax = table_videos11_rename.plot(kind='bar',title='Video Categories by their Count',color='green',figsize=(10,5))
ax.set_xlabel('Category')
ax.set_ylabel('Count')

plt.tight_layout()
plt.savefig('video_trends_us.png', dpi=300)
```

```python
# plot video counts in each category for British videos

table_videos5_cat_id = table_video5.groupby('categoryId').count()['title']
table_videos5_rename = table_videos5_cat_id.rename(category_id_map)

ax = table_videos5_rename.plot(kind='bar',title='Video Categories by their Count',color='green',figsize=(10,5))
ax.set_xlabel('Category')
ax.set_ylabel('Count')

plt.tight_layout()
plt.savefig('video_trends_british.png', dpi=300)
```

```python
# Show a word cloud of all keywords in tags

from wordcloud import STOPWORDS,WordCloud

table_videos.head()

tags = table_videos['tags'].map(lambda x:x.lower().split('|')).values

splt_tags = ' '.join(table_videos['tags'])

wordcloud = WordCloud(width=1000, height=500).generate(' '.join(splt_tags.lower().split('|')))

plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')

plt.tight_layout()
plt.savefig('tags_word_cloud.png', dpi=300)
```

```python
# Determine all relevant keywords for certain topic and identify the trends through another perspective

videos_tech = table_videos[table_videos['tags'].str.contains('Apple' or 'Iphone' or 'Aviation' or 'airplane' or 'Boeing' or 'Airbus' or 'Technology')].sort_values(by='view_count', ascending=False)
```

```python
# plot hte view counts for the top ten videos in each category; here only 3 topic are listed, but you can do for other topic of your interest

ax = videos_tech.head(10).plot(kind='bar',x='channelTitle',y='view_count',title='Top 10 channels in Technology')
ax.set_xlabel('channelTitle')
ax.set_ylabel('Views(in m)')

plt.tight_layout()
plt.savefig('video_trends_technology.png', dpi=300)
```
